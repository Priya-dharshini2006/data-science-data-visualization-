```python
import pandas as pd
import pandas as pd
from google.colab import files
uploaded = files.upload()
gold_dataset=pd.read_csv("Gold vs USDINR.csv")
gold_dataset.head()
```

```python
gold_dataset.head()
```

```python
gold_dataset['Goldrate']=gold_dataset['Goldrate'].replace('â','',regex=True).replace(',','',regex=True)
gold_dataset.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.pairplot(gold_dataset)
plt.show()
```

```python
sns.boxplot(gold_dataset)
plt.show()
```

```python
gold_dataset['Goldrate'] = gold_dataset['Goldrate'].astype(str).str.replace(r'[^\d.]','',regex=True)
gold_dataset['Goldrate'] = pd.to_numeric(gold_dataset['Goldrate'])
sns.regplot(x='Goldrate', y='USD_INR', data=gold_dataset)
plt.show()
```

```python
x=gold_dataset['USD_INR']
y=gold_dataset['Goldrate']
x
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
x_train
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train_scaled = scaler.fit_transform(x_train.values.reshape(-1,1))
x_test_scaled = scaler.transform(x_test.values.reshape(-1,1))
print(x_train_scaled)
print(x_test_scaled)
```

```python
x_train_scaled,x_test_scaled
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train_scaled,y_train)
```

```python
regressor.get_params()
```

```python
regressor.coef_
```

```python
regressor.intercept_
```

```python
m=regressor.coef_
b=regressor.intercept_
m,b
```

```python
x_train_predict=regressor.predict(x_train_scaled)
plt.scatter(x_train,y_train)
plt.plot(x_train,x_train_predict,color='red')
plt.xlabel('USDINR')
plt.ylabel('Goldrate')
plt.title('USDINR vs Goldrate')
plt.show()

```

```python
x_test_predict=regressor.predict(x_test_scaled)
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test,x_test_predict)
```
